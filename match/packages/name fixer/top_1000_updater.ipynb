```python
for i in range(len(df)):
    trigger = df.at[i, 'name']
    var = False
    for t in lines:
        if trigger in t:
            var = False
        else:
            var = True
    if var == True:
        
        replace = df.at[i, 'replace']
        line = f"  - regex: {trigger}\n    replace: {replace}\n    word: true\n"
        lines.append(line)
```
