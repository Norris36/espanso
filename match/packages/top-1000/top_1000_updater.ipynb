```python
import itertools
import itertools
import pandas as pd
from tqdm import tqdm
```

```python
def generate_permutations(word):
    """
    Generates all possible permutations of a given word, up to a maximum length of 4.
    """
    if len(word) > 4:
        return []
    else:
        permutations = [''.join(p) for p in itertools.permutations(word)]
    return permutations

def generate_permutations(word):
    """
    Generates all possible permutations of a given word, including variations with missing letters.
    """
    if len(word) <= 1:
        return [word]
    else:
        permutations = []
        for i in range(len(word)):
            sub_permutations = generate_permutations(word[:i] + word[i+1:])
            for p in sub_permutations:
                permutations.append(word[i] + p)
                permutations.append(p)
        return permutations
    



def generate_permutations(word):
    """
    Generates all possible permutations of a given word based on its length.
    """
    if len(word) <= 4:
        # Generate all possible permutations of the word
        permutations = [''.join(p) for p in itertools.permutations(word)]
    else:
        permutations = []
        # Generate all possible permutations of the word missing one character
        for i in range(len(word)):
            permutations.append(word[:i] + word[i + 1:])
    return permutations
```

```python
df = pd.read_csv(filepath_or_buffer='dictionary.csv', delimiter=";", index_col=0)
df = df[(df.danish.str.len()>1)]
df = df[(df.english.str.len()>1)]
english = df[['english']].copy()
danish = df[['danish']].copy()

english.rename(columns={'english':'correct_spelling'}, inplace=True)
danish.rename(columns={'danish':'correct_spelling'}, inplace=True)

# english['misspellings'] = english.correct_spelling.str.len()
# danish['misspellings'] = danish.correct_spelling.str.len()

english['misspellings'] = ''
danish['misspellings'] = ''
```

```python
danish.misspellings.describe()
```

```python
for word in tqdm(english.correct_spelling, desc='This is the process for the english words'):
    dictionary = english.correct_spelling.to_list() + danish.correct_spelling.to_list() + english.misspellings.to_list() + danish.misspellings.to_list() 
    
    new = generate_permutations(word)

    tmp = []
    for each in new:
        if each in dictionary:
            pass
        else:
             tmp.append(each)

    new = tmp

    english.loc[english.correct_spelling == word, 'misspellings'] = str(new)

for word in tqdm(danish.correct_spelling, desc='This is the process for the danish words'):
    dictionary = english.correct_spelling.to_list() + danish.correct_spelling.to_list() + english.misspellings.to_list() + danish.misspellings.to_list() 
    
    new = generate_permutations(word)

    tmp = []
    for each in new:
        if each in dictionary:
            pass
        else:
             tmp.append(each)

    new = tmp

    danish.loc[danish.correct_spelling == word, 'misspellings'] = str(new)
```
